```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import pandas as pd
import re

text_file=[]
df2=pd.read_csv('T_mygeometryv1_stf.txt', header = None, dtype=float)
count=0
for (root,dirs,files) in os.walk('.\\',topdown=True):
  for x in files:
    if x.endswith(".txt") == True:
      count+=1
      pattern = ".txt"
      new_text = re.sub(pattern, '', x)
      data = pd.read_csv(str(x), header = None, dtype=float)
      text_file.append(str(new_text))
      df2[count]=data[1]

d=df2.set_index(0)

plt.figure()
plt.plot(d)
plt.axhline(y=0.9, color='lightgray', linestyle='--', label="Y = 0.9")
plt.legend(text_file)
plt.title('Comparison Python and Lumerical')
plt.ylabel('Transmission')
plt.xlabel('Wavelengths(m)')
plt.ylim(0,1.0)
plt.grid(True)
plt.savefig('comparison.png',dpi=300)
plt.show()
```
